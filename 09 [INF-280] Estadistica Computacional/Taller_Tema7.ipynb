Primero se importaran las librerias que se necesitan.

```python
import pandas as pd
import numpy as np
from numpy.random import seed
from numpy.random import normal
from numpy.random import choice
from numpy.random import randint
import matplotlib.pyplot as plt
import seaborn as sn
from scipy.stats import t, norm
from statsmodels.stats.weightstats import zconfint
```

```python
def IC_media_T(muestra, alpha=0.05):
    n = len(muestra)
    media_muestra = np.mean(muestra)
    desv_est_muestra = np.std(muestra)
    conf = 1 - alpha
    t_alpha = t.ppf(1 - alpha/2, df=n-1)
    ic_inf = media_muestra - t_alpha * desv_est_muestra/np.sqrt(n)
    ic_sup = media_muestra + t_alpha * desv_est_muestra/np.sqrt(n)
    return (ic_inf, ic_sup)

def IC_media_Norm(muestra, alpha=0.05, desv_est_real=None):
    n = len(muestra)
    media_muestra = np.mean(muestra)
    if desv_est_real is None:
        desv_est_real = np.std(muestra)
    conf = 1 - alpha
    z_alpha = norm.ppf(1 - alpha/2)
    ic_inf = media_muestra - z_alpha * desv_est_real/np.sqrt(n)
    ic_sup = media_muestra + z_alpha * desv_est_real/np.sqrt(n)
    return (ic_inf, ic_sup)

def IC_difmed_Norm(muestraX, muestraY, alpha=0.05):
    nX     = len(muestraX)
    stdX   = np.std(muestraX)
    mediaX = np.mean(muestraX)
    nY     = len(muestraY)
    stdY   = np.std(muestraY)
    mediaY = np.mean(muestraY)
    z_alpha = norm.ppf(1 - alpha/2)
    ic_inf  = (mediaX - mediaY) - z_alpha * np.sqrt((stdX**2)/nX + (stdY**2)/nY)
    ic_sup  = (mediaX - mediaY) + z_alpha * np.sqrt((stdX**2)/nX + (stdY**2)/nY)
    return (ic_inf, ic_sup)
```

## 2. L

Se nos pide calcular **intervalos de confianza para proporciones** de arresto para cada raza, por lo tanto, la variable aleatoria con la que se esta trabajando es una Bernoulli (esta arrestado o no).

La cantidad pivotal corresponde a:
\begin{eqnarray}
Z = \frac{\widehat{p}-p}{\sqrt{\frac{\widehat{p}(1-\widehat{p})}{n}}}
\end{eqnarray}

```python
races = list(df_sqf2019['SUSPECT_RACE_DESCRIPTION'].value_counts().keys())

l_ic_2 = []
z = 1.96 # Obtenido manualmente en tablas

for raza in races:
    total = df_sqf2019.query(f"SUSPECT_RACE_DESCRIPTION == '{raza}'").count()[1]
    total_arrest = df_sqf2019.query(f"SUSPECT_RACE_DESCRIPTION == '{raza}'").query("SUSPECT_ARRESTED_FLAG == 'Y'").count()[1]
    p_estim = total_arrest/total
    print(f"Raza: {raza} - Total quest: {total} - Total arrest: {total_arrest} - Prop: {p_estim}")
    
    lim_sup = p_estim + z * np.sqrt((p_estim * (1 - p_estim))/(n_df_sqf2019))
    lim_inf = p_estim - z * np.sqrt((p_estim * (1 - p_estim))/(n_df_sqf2019))
    ic = (lim_inf, lim_sup)
    l_ic_2.append(ic)
    
    print(f"[{lim_inf}, {lim_sup}]\n")
```

```python
f = plt.figure(figsize=(7,4))
ax = f.add_axes([0.1,0.1,0.9,0.9])

for i in range(len(l_ic_2)):
    ax.plot([i,i],l_ic_2[i],'o-')

ax.legend(races)
plt.ylabel('Intervalos de confianza')
plt.show()
```

Se puede ver que el intervalo de confianza para los indio-americanos es mayor que el resto, pero esto se puede deber a la poca cantidad de personas que se han cuestionado, en este caso: solo 8.

Sin embargo, la mayor cantidad de gente cuestionada son los de raza negra, con 7981 personas, de las cuales 2499 fueron arrestadas. La mayor cantidad siguiente son los de raza blanca hispana, con 2742 personas, de las cuales 910 fueron arrestadas. Notando las proporciones entre ambas, la segunda es mayor a la primera.
